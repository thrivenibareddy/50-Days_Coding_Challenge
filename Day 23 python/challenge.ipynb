```python
# Input data (single string)
data = "Ride101-Ramesh-Late Arrival-Bangalore, Ride102-Suresh-rude behavior-Hyderabad, Ride103-Mahesh-Late arrival-Bangalore"
```

```python
# Task 1: Clean the Data
# Convert entire string to lowercase
data = data.lower()
print("Data:",data)
```

```python
# Task 2: Count Total Complaints
# Each record separated by comma
total_complaints = data.count(",") + 1
print("Total Complaints:", total_complaints)
```

```python
# Task 3: Count 'Late Arrival' Issues
late_arrival_count = data.count("late arrival")
print("Late Arrival Issues:", late_arrival_count)
```

```python
# Task 4: Count complaints from Bangalore
bangalore_count = data.count("bangalore")
print("Bangalore Complaints:", bangalore_count)
```

```python
# Task 5: Extract First Driver Name
first_record = data.split(",")[0]
driver_name = first_record.split("-")[1]
print("First Driver Name:", driver_name)


```

```python
# Task 6: Replace Issue Names (Standardization)
standardized_data = data.replace("rude behavior", "rude_behavior")
standardized_data = standardized_data.replace("late arrival", "late_arrival")
print("Standardized Data:", standardized_data)
```
