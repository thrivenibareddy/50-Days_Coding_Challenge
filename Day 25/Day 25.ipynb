```python
students = ("sunitha", "RAVI", "aNu")

proper_names = tuple(name.capitalize() for name in students)

print(proper_names)
```

```python
subjects = ("Math", "Science", "English", "Math", "Science")

# Total subjects
total = len(subjects)

# Unique subjects
unique = set(subjects)

print("Total Subjects:", total)
print("Unique Subjects:", unique)
```

```python
feedback = ("  good course  ", " excellent ", "average ")

clean_feedback = tuple(f.strip() for f in feedback)

print(clean_feedback)
```

```python
names = ("Anu", "Ravi", "John")
courses = ("Python", "SQL", "Power BI")

combined = tuple(names[i] + "-" + courses[i] for i in range(len(names)))

print(combined)

```

```python
emails = ("anu@gmail.com", "ravi@yahoo.com", "john@edu.com")

domains = tuple(email.split("@")[1] for email in emails)

print(domains)
```
