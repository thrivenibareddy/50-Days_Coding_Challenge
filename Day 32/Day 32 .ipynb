```python
#1. Even or Odd Checker
num = int(input("Enter a number: "))

if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
#Positive, Negative or Zero
num = int(input("Enter a number: "))

if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")
```

```python
#Sum of First N Numbers
n = int(input("Enter N: "))

i = 1
total = 0

while i <= n:
    total += i
    i += 1

print("Sum:", total)
```

```python
#5. Password Retry System
correct_password = "admin123"
attempts = 0

while attempts < 3:
    password = input("Enter password: ")

    if password == correct_password:
        print("Access Granted")
        break
    else:
        print("Wrong Password")
        attempts += 1

if attempts == 3:
    print("Access Denied")
```

```python
#6. Multiplication Table
num = int(input("Enter a number: "))

i = 1
while i <= 10:
    print(num, "x", i, "=", num * i)
    i += 1
```
