```python
students = {
    "Asha": {"Math": 78, "Science": 85, "English": 67},
    "Ravi": {"Math": 35, "Science": 40, "English": 50},
    "John": {"Math": 90, "Science": 92, "English": 88}
}
totals = {}
averages = {}
```

```python

#Calculate Total & Average
for name, subjects in students.items():
    total = sum(subjects.values())
    average = total / len(subjects)

    totals[name] = total
    averages[name] = average

    print(f"{name} → Total: {total}, Average: {average:.2f}")
```

```python

#Identify Top Scorer
top_scorer = max(totals, key=totals.get)
print("Top Scorer:", top_scorer)
```

```python
#Find Failed Students (marks < 40)
failed_students = []

for name, subjects in students.items():
    for mark in subjects.values():
        if mark < 40:
            failed_students.append(name)
            break

print("Failed Students:", failed_students)
```
