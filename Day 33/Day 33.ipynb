```python

#Number guessing game
secret_number = 7

while True:
    guess = int(input("Enter your guess: "))

    if guess < secret_number:
        print("Too Low")
    elif guess > secret_number:
        print("Too High")
    else:
        print("You Won!")
        break
```

```python
#reverse a number
num = int(input("Enter a number: "))
reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10

print("Reversed number:", reverse)
```

```python
#count digits in number
num = int(input("Enter a number: "))
count = 0

while num > 0:
    num = num // 10
    count += 1

print("Total digits:", count)
```

```python
#Sales Threshold Checker
sales = [1200, 800, 1500, 400, 2000]

count = 0

for s in sales:
    if s > 1000:
        print("High Sale:", s)
        count += 1

print("Total high sales:", count)
```

```python
#Continuous Data Entry System
total = 0

while True:
    num = int(input("Enter number (0 to stop): "))

    if num == 0:
        break

    total += num

print("Total sum:", total)
```
